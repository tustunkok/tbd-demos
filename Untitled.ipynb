```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import classification_report
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from keras.layers import Conv2D, MaxPooling2D, Dense
from keras.layers import Dropout, Flatten, Input
from keras.models import Model
```

```python
df = pd.read_csv("mnist.csv")
df.head()
```

```python
onehot_enc = OneHotEncoder(sparse=False)
X = df.drop("label", axis=1).values
y = onehot_enc.fit_transform(df["label"].values.reshape(-1, 1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, \
    test_size=0.14, random_state=42)
```

```python
import keras.backend as K
K.clear_session()

inputs = Input(shape=(28, 28, 1))
cnns = Conv2D(10, kernel_size=9, activation="relu")(inputs)
cnns = MaxPooling2D(pool_size=(2,2))(cnns)
cnns = Conv2D(10, kernel_size=9, activation="relu")(cnns)

cnns = Flatten()(cnns)

deepnn = Dense(20, activation="relu")(cnns)
deepnn = Dropout(0.5)(deepnn)
outputs = Dense(10, activation="softmax")(deepnn)

complete_model = Model(inputs=inputs, outputs=outputs)
complete_model.summary()
```

```python
complete_model.compile(loss="categorical_crossentropy", \
                      optimizer="adam",
                      metrics=["accuracy"])
```

```python
complete_model.fit(X_train.reshape(-1, 28, 28, 1), y_train, \
                  batch_size=128, epochs=5)
```

```python
partial_model = Model(inputs=inputs, outputs=cnns)
partial_model.summary()
```

```python
rf_train_data = partial_model.predict(X_train.reshape(-1, 28, 28, 1))
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(rf_train_data, y_train)
```

```python
rf_test_data = partial_model.predict(X_test.reshape(-1, 28, 28, 1))
```

```python
predictions = rf_model.predict(rf_test_data)
```

```python
y_test = onehot_enc.inverse_transform(y_test)
predictions = onehot_enc.inverse_transform(predictions)
```

```python
print(classification_report(y_test, predictions))
```
